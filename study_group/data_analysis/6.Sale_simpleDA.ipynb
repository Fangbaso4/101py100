```python
import pandas as pd
import numpy as np
```

```python
# 利用 pandas 读取 task6_1.csv 文件中的数据，建立 DataFrame
df = pd.read_csv('task6_1.csv')
df
```

```python
# 利用 pivot_table 统计该企业每个人每天销售商品 A 和 B 的金额之和
pd.pivot_table(df, index=['id'], columns=['date'], values=['sales'], aggfunc=np.sum)
```

```python
# 请将上面的统计数据再增加一列，统计出每个人的销售总金额
pd.pivot_table(df, index=['id'], columns=['date'], values=['sales'], aggfunc=np.sum, margins=True)
```

```python
# 利用 pandas 读取 task6_2.csv 文件中的数据，建立 DataFrame
df2 = pd.read_csv('task6_2.csv')
df2
```

```python
# 请以性别为分组依据，查看男女抽烟平均年龄和身高
pd.pivot_table(df2, index=['sex'], values=['age', 'height'], aggfunc=np.mean)
```

```python
# 利用交叉表知识，统计各个性别抽烟的人数
pd.crosstab(df2['sex'], df2['smoke'])
```

```python
# 利用交叉表知识，统计各个年龄段抽烟人情况
a,b = pd.cut(x=df2["age"],bins=3,right=True,retbins=True)

df2['age_cut'] = pd.cut(df2.age, bins=b)
df2
```

```python
pd.crosstab(df2.age_cut, df2.smoke, margins=True)
```
