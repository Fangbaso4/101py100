```python
import numpy as np
import pandas as pd
```

```python
data = {
    'marketcap': [449, 371, 237, 21313, 1369, 823], 
    'pe': [8.31, 15.36, 16.01, 7.16, 7.59, 6.3],
    'code': ['600926', '002958', '601128', '601398', '601229', '600919']
}
df = pd.DataFrame(data, index= ['杭州银行', '青农商行', '常熟银行', '工商银行', '上海银行', '江苏银行'])
df
```

```python
# 选出市值低于 2000 亿的所有公司
df.loc[df.marketcap<2000]
```

```python
# 选出市值 < 2000亿，并且市盈率 < 10 的所有公司（
df.loc[(df.marketcap<2000) & (df.pe<10.0)]
```

```python
# 假设其中一家公司股票的每日收盘价数据，如下
import numpy as np
dr = pd.date_range(start='2019-01-02', periods=100)
data = np.random.randn(100).cumsum()
close = data - np.min(data)
df = pd.DataFrame({"close": close}, index=dr)
df
```

```python
# 请计算收盘价的五日平均，并新增一列存储
df['5Davg'] = df.rolling(window=5).mean()
df
```

```python
# 创建 DataFrame 存储数据（
data = np.arange(1,17)
data = data.reshape(4,4)
df = pd.DataFrame(data, index=['a', 'b', 'c', 'd'], columns=['第1列', '第2列', '第3列', '第4列'])
df
```

```python
# 请利用 apply 函数，统计出每一行的总和
df.loc['row_sum'] = df.apply(lambda x: x.sum())
df
```

```python
# 请生成第 5 列数据，使其满足：当第 3 列数据 > 10 时，第 5 列数据等于第 1 、第 2 列数据之和；否则第 5 列数据等于第 3 列数据
df['第五列'] = df.apply(lambda x: x[0]+x[1] if x[2]>10 else x[2], axis=1)
df
```
